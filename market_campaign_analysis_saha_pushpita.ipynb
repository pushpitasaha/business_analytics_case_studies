#### Case 2: Marketing Campaign Analysis (Question 4)

 You are a data analyst at DigitalStream, a digital marketing agency. The agency recently ran
 multiple online advertising campaigns for various clients. Your task is to analyze the
 performance of these campaigns to determine which had the highest engagement rate and
 suggest an optimization strategy.

 image.png

######Question 4:
 ● Calculate the engagement rate for each campaign using the formula `(clicks/impressions) * 100` and update each dictionary with its respective engagement rate.

 - a. Write a function `best_campaign` to identify and print the campaign with the
 highest engagement rate.
 - b. Utilize loops and conditional statements to analyze and update the campaign data.

```python
# Assumptions made:
# In the input list of campaigns dictionaries, every campaign data is saved as key:value pairs in dict
# engagement rate has to be created as another k:v pair for each of them
# List of campaign data (auto-generated with AI instead of manual typing)
campaigns = [
    {'name': 'Campaign A', 'clicks': 120, 'impressions': 3000},
    {'name': 'Campaign B', 'clicks': 150, 'impressions': 3500},
    {'name': 'Campaign C', 'clicks': 90, 'impressions': 2000},
    {'name': 'Campaign D', 'clicks': 200, 'impressions': 4000},
]
```

```python
#Following code modularity I am separating concerns into small tasks

# First, calculate engagement rate
# and update each campaign
def calculate_engagement_rate(campaigns):
    for campaign in campaigns: # for each dict in given list
        engagement_rate = (campaign['clicks'] / campaign['impressions']) * 100 # given formula to calculate eng rate
        campaign['engagement_rate'] = engagement_rate  # update campaign with engagement rate: value pairs, a new dict entry
```

```python
# Second, find the campaign with the highest engagement rate
def find_highest_engagement_campaign(campaigns):
    highest_engagement_campaign = None
    highest_engagement_rate = 0 #init with zero, any new value found will be higher

    for campaign in campaigns:
        if campaign['engagement_rate'] > highest_engagement_rate: # check if eng rate is higher
            highest_engagement_rate = campaign['engagement_rate'] # update highest rate
            highest_engagement_campaign = campaign # update highest campaign name

    return highest_engagement_campaign # push to global
```

```python
# Third, print the best campaign
def print_best_campaign(campaign):
    if campaign:
        print("Campaign with the highest engagement rate:")
        print(f"Name: {campaign['name']}, Engagement Rate: {campaign['engagement_rate']:.2f}%") # name and rate will be printed
    else:
        print("No campaign data available.")
```

```python
# Finally, I am implementing a main function to put every part in order
def best_campaign(campaigns):
    calculate_engagement_rate(campaigns) # 1. eng rate calc and update vals
    highest_campaign = find_highest_engagement_campaign(campaigns) # 2. find highest rate and save that value
    print_best_campaign(highest_campaign) # 3. print highest rate with that key
```

```python
best_campaign(campaigns) #run the logic
```
